# Data

**Exploring** **the** **Data**

```python
import pandas as pd

data = pd.read_csv("WineQT.csv")
data
```

**Looking** **at** **components** **of** **the** **Data**

```python
data.info()
```

```python
import missingno as msno

_ = msno.bar(data, figsize=(10, 5), color='#FF281B')
```

**Fixing** **the** **Dataframe**

```python
data = data.drop(columns=['Id'])
data
```

**Exploratory** **Data** **Analysis**

**Exploring** **the** **variables**

```python
data.describe().T.style.background_gradient(cmap='Reds').set_precision(3)
```

**Correlation**

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

corrMatrix = data.corr()
plt.rcParams["figure.figsize"] = (12,10)
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

**Boxplot**

```python
plt.style.use('ggplot')
plt.figure(figsize=(20,60))

for i,j in enumerate(data.columns):
    plt.subplot(12, 3, i+1)
    fig = data.boxplot(column=j)
    fig.set_title('')
    fig.set_ylabel(j)

plt.show()
```

**Scatterplot**

```python
from pandas.plotting import scatter_matrix

_ = scatter_matrix(data.drop(columns=['quality']), c=data.quality,figsize=(20,20), s=50)
```

**Histogram**

```python
fig = plt.figure(figsize = (20,16))
for i,j in enumerate(data.columns):
    plt.subplot(4,3,i+1)
    sns.distplot(data[j], color='r')
plt.show()
```

**Distribution** **of** **the** **Target** **Variable**

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, specs=[[{"type": "pie"}, {"type": "bar"}]])
fig.add_trace(go.Pie(values = data.quality.value_counts(), labels = data.quality.value_counts().index, domain=dict(x=[0, 0.5]), 
                     marker=dict(colors=['#FC4F4F', '#FFBC80', '#FF9F45', '#F76E11', '#CD104D', '#820000']), hole=.3), row=1, col=1)
fig.add_trace(go.Bar(x=data.quality.value_counts().index, y=data.quality.value_counts(), name=''), row=1, col=2)
fig.update_layout(showlegend=False)
fig.show()
```

**Comparison** **Plot**

```python
plt.style.use('seaborn')
fig,ax = plt.subplots(4,3,figsize=(20,30))
for i,j in enumerate(list(data.columns)[:-1]):
    sns.boxplot(x = data.quality,y = data[j],ax = ax[i//3][i%3])
plt.delaxes(ax[3,2])
```

**Skewness**

```python
plt.figure(figsize=(12,7))
skewness = data.skew().sort_values(ascending=False)
sns.barplot(x=skewness, y=skewness.index, palette="seismic_r")
plt.axvline(x=1, color='Red', linestyle='--')
plt.xlabel('Skewness')
plt.ylabel('Variables')
plt.show()
```
